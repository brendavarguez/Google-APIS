```python
import pandas as pd
import requests
import json
```

The Hoyodecrimen REST API (https://hoyodecrimen.com/api/) allows to query crime data about Mexico City. 

Complete the following tasks:

* Obtain a **list** of all crime types in the Database.
* Obtain all quadrants in the Database and store them into a table with the following columns:
    * cuadrante
    * cve_mun
    * municipio

```python
# get the url about crime data in Mexico City
response = requests.get('https://hoyodecrimen.com/api/v1/crimes')
data = json.loads(response.content)
print(data)
```

```python
# pass the json file into a data frame to print it as a table
crimen = pd.read_json(json.dumps(data['rows']), orient = 'records')
crimen
```

```python
# Obtaining all quadrants in the Database as a json file
response1 = requests.get("https://hoyodecrimen.com/api/v1/cuadrantes")
data1 = json.loads(response1.content)
print(data1)
```

```python
# and store quadrants into a table with cuadrant, cve_num, municipio and sector columns
cuadrantes = pd.read_json(json.dumps(data1['rows']), orient = 'records')
cuadrantes
```

* Create a function to obtain all the crimes within a certain distance given the geographic coordinates of a point (lat/lon). Create a table locating the crimes near to the Zocalo (19.433735, -99.133334):
    * crime
    * date
    * hour
    * lat
    * lon

```python
def find_crimes():
    
    """ This function will obtain all crimes within a distance of 500 meters.
        ARGS: Does not receives any argumen.
        RETURNS: A json file with the crime, data, hour and location(lat, lon).
    """
    
    crimes_url = 'https://hoyodecrimen.com/api/v1/latlong/crimes/{}/coords/{}/{}/distance/{}'.format('all', '-99.133334','19.433735','500')
    res = requests.get(crimes_url)
    results = json.loads(res.content)
    return results
```

```python
all_1 = find_crimes()
all_1
```

```python
# pass the json file into a data frame to make the data easier to read.
all_crimes = pd.read_json(json.dumps(all_1['rows']), orient = 'records')
all_crimes
```

* Obtain the top ranked quadrants the highest crime counts in 2016, 2017, 2018. Create a unique table including:
    * year
    * crime
    * quadrant
    * difference
    * population
    * rank

```python
def count_crimes(date):
    
    """ This function will read a json files with the ranked quadrats with highest crime counts between 2016-2018.
        ARGS: The year we are interested to analyze. 
        RETURNS: A json file with the crime, start and end period of the crime, sector, cuadrant, rank, etc.
    """
    
    ranked_url = 'https://hoyodecrimen.com/api/v1/cuadrantes/crimes/all/top/counts?start_date={}-01&end_date={}-01'.format(date, (date + 1))
    res3 = requests.get(ranked_url)
    results3 = json.loads(res3.content)
    return results3
```

```python
# call the function to obtain the data between 2016-2018
count_crimes_2016 = count_crimes(2016)
count_crimes_2017 = count_crimes(2017)
count_crimes_2018 = count_crimes(2018)
```

```python
# pass the data into a data frame to make it easier to read by reading the rows ehre the columns of the data
# frame are held and with recors orient.
count_crimes_2016 = pd.read_json(json.dumps(count_crimes_2016['rows']), orient = 'records')
count_crimes_2017 = pd.read_json(json.dumps(count_crimes_2017['rows']), orient = 'records')
count_crimes_2018 = pd.read_json(json.dumps(count_crimes_2018['rows']), orient = 'records')
all_crimes = pd.concat([count_crimes_2016, count_crimes_2017, count_crimes_2018])
all_crimes['year'] = all_crimes['start_period'].str.slice(start = 0, stop = 4)
```

```python
# drop the unecessary columns
all_crimes.drop(['count', 'end_period', 'sector', 'start_period'], axis=1)
```
